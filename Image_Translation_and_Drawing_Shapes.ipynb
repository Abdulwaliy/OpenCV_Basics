```python
# import the necessary packages
from matplotlib import pyplot as plt
import numpy as np
import imutils
import cv2
```

```python
def plt_imshow(title, image):
	# convert the image frame BGR to RGB color space and display it
	image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
	plt.imshow(image)
	plt.title(title)
	plt.grid(False)
	plt.show()
```

```python
#Read file from google drive
from google.colab import drive
drive.mount('/content/drive')
image = cv2.imread("//content/drive/MyDrive/Open CV Pictures/archive/data/opencv-logo.png")
plt_imshow("open cv logo", image)
```

```python
# @title Image Translation

plt_imshow("open cv logo", image)
```

```python
# Use imutils method
translated_image = imutils.translate(image, 50, 100) # shift left by 50 and shift down by 100
plt_imshow("translated_image", translated_image)
```

```python
# shift the image 25 pixels to the right and 50 pixels down
M = np.float32([[1, 0, 25], [0, 1, 50]]) # create the shifting canvas
shifted = cv2.warpAffine(image, M, (image.shape[1], image.shape[0])) #more or less like putting the image into the shifting canvas
plt_imshow("Shifted Down and Right", shifted)
```

```python
M = np.float32([[1, 0, -50], [0, 1, -90]])
shifted = cv2.warpAffine(image, M, (image.shape[1], image.shape[0]))
plt_imshow("Shifted Up and Left", shifted)
```

```python
# @title Drawing Shapes

# First create a canvas
canvas = np.zeros((400, 400, 3), dtype = "uint8" ) #pixels are represented by unsigned 8bit

#plt_imshow("our canvas",canvas)

#b, g, r
#g, b, r
red = (0,0,255)
cv2.line(canvas, (0,0), (400,400), red )
plt_imshow("our canvas",canvas)
```

```python
blue = (255, 0 ,0)
cv2.rectangle(canvas, (200,200), (40,80), blue, 5 )
plt_imshow("our canvas",canvas)
```

```python
cv2.circle(canvas, (250, 300), 100, blue)
plt_imshow("our canvas",canvas)
```
