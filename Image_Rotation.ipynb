```python
# import the necessary packages
from matplotlib import pyplot as plt
import numpy as np
import imutils
import cv2
```

```python
def plt_imshow(title, image):
	# convert the image frame BGR to RGB color space and display it
	image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
	plt.imshow(image)
	plt.title(title)
	plt.grid(False)
	plt.show()
```

```python
#Read file from google drive
from google.colab import drive
drive.mount('/content/drive')
image = cv2.imread("//content/drive/MyDrive/Open CV Pictures/archive/data/opencv-logo.png")
plt_imshow("open cv logo", image)
```

```python
rotate = imutils.rotate(image, -50)
```

```python
plt_imshow("open cv logo_rotate", rotate)
```

```python
rotate = imutils.rotate_bound(image, 50)
plt_imshow("open cv logo_rotate", rotate)
```
